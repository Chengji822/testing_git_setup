# 0. Importing and Observing Data

```python
df = pd.read_csv('U.S._Chronic_Disease_Indicators__CDI___2023_Release_20240923.csv')
df
```

# 1. Select Relevant Columns

```python
# find unique values that include "binge drinking" and "poverty"
df['Question'].unique()
```

```python
# select data that is about: 
# Binge drinking prevalence among adults aged >= 18 years (overall, males, females), in crude prevalence percentages.
binge_drinking_crude = df[(df['Question'] == 'Binge drinking prevalence among adults aged >= 18 years')
                         &(df['DataValueTypeID'] == 'CRDPREV')
                         ]
binge_drinking_crude_OveralFM = binge_drinking_crude[(binge_drinking_crude['StratificationCategoryID1'] == 'OVERALL')
                                                    |(binge_drinking_crude['StratificationCategoryID1'] == 'GENDER')
                                                    ]
binge_drinking_crude_OveralFM
```

```python
# select data that is about: 
# Poverty prevalence, Crude Prevalence in Percent (overall only).
poverty_crude_overall = df[(df['Question'] == 'Poverty') 
                         & (df['DataValueType'] == 'Crude Prevalence')
                         & (df['StratificationCategoryID1'] == 'OVERALL')
                          ]
poverty_crude_overall2 = poverty_crude_overall.copy()
poverty_crude_overall2['StratificationID1'] = 'poverty'
poverty_crude_overall2
```

```python
# concat the two dataframe into one df
binge_poverty = pd.concat([binge_drinking_crude_OveralFM, poverty_crude_overall2], axis=0)
binge_poverty
```

# 2. Reshape the Data

```python
# renaming parts of columns
binge_poverty = binge_poverty.rename(columns={
    'YearEnd': 'year',
    'LocationAbbr': 'stateabb',
    'LocationDesc': 'state',
})
binge_poverty['DataValue'] = binge_poverty['DataValue'].astype(float)


# Convert the dataset from long format to wide format using appropriate pandas commands. 
df_wide = binge_poverty.pivot(index=['year', 'stateabb', 'state'], columns = 'StratificationID1', values = 'DataValue')
df_wide
```

```python
# renaming columns
df_wide = df_wide.rename(columns={
    'GENF': 'binge_female',
    'GENM': 'binge_male',
    'OVR': 'binge_all',
})
df_wide
```

# 5.Data Transformation and Analysis

* 5.1 Top 10 States for Binge Drinking

```python
# For the most recent year in the dataset
df_wide2 = df_wide.copy()
df_wide2 = df_wide2.reset_index()
max_year = df_wide2['year'].max()
df_wide_max_year = df_wide2[df_wide2['year'] == max_year]

# create a table showing the top 10 U.S. states with the highest overall binge drinking prevalence.
overall_BD_maxYear = df_wide_max_year.sort_values(by='binge_all', ascending=False)
top_ten = overall_BD_maxYear.head(10)
Top10States_OverallBD_maxYear = top_ten[['year','state','binge_all']]
Top10States_OverallBD_maxYear.index = [i for i in range(1,11)]
Top10State_table = pd.DataFrame(Top10States_OverallBD_maxYear)
Top10State_table
```

* 5.2 Annual Growth Rates in Binge Drinking

```python
df_wide3 = df_wide.copy()
df_wide3 = df_wide3.reset_index()
bing_all = df_wide3[['year', 'state','binge_all']]

# Drop rows with missing binge_all values to ensure accurate calculations
bing_all_clean = bing_all.dropna(subset=['binge_all'])

# Sort the data by 'state' and 'year' to ensure pct_change works correctly
bing_all_clean = bing_all_clean.sort_values(['state', 'year'])

# Group by 'state' and calculate the year-over-year percentage change using pct_change
bing_all_clean['pct_change'] = bing_all_clean.groupby('state')['binge_all'].pct_change()

# Now, calculate the average of these percentage changes for each state (AAGR approximation)
aagr_by_state = bing_all_clean.groupby('state')['pct_change'].mean().reset_index()
aagr_by_state.columns = ['state', 'avg_annual_growth_rate']

# top 5
sorted_aagr = aagr_by_state.sort_values(by='avg_annual_growth_rate', ascending=False)
top_5 = sorted_aagr.head(5)
bottom_5 = sorted_aagr.tail(5)

# create table
top_5.index = [i for i in range(1,6)]
bottom_5.index = [i for i in range(1,6)]

Top5State_table = pd.DataFrame(top_5)
bottom5State_table = pd.DataFrame(bottom_5)

Top5State_table
```
